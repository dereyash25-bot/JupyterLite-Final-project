***Exercise 9: Convert minutes to hours***

```python
def minutes_to_hours(minutes):
    hours = minutes // 60
    mins = minutes % 60
    return f"{minutes} minutes = {hours} hour(s) and {mins} minute(s)"

# Example
minutes_to_hours(135)
```
